```python
#1. Crear máscara para identificar nulos en 'currency-code'
mascara_nulos = df['currency-code'].isna()

# 2. Filtrar el dataframe para quedarte solo con los registros nulos
registros_nulos = df[mascara_nulos]

# 3. Agrupar por 'country' y 'continent' para contar nulos por grupo
patron = registros_nulos.groupby(['country', 'continent']).size().reset_index(name='cantidad_nulos')

# 4. Mostrar los países y continentes con nulos en 'currency-code'
print("Países y continentes con nulos en currency-code:")
print(patron)
```

```python
# 1. Agrupar por 'country' y 'currency-code' para contar nulos por grupo
patron = df.groupby(['country', 'currency-code']).size().reset_index(name='cantidad_nulos')

# 2. Mostrar los países y continentes con nulos en 'currency-code'
print("Países y continentes con nulos en currency-code:")
print(patron)
```
