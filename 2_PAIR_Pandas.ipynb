Momento de trabajar con Series. Contesta a las siguientes preguntas:

3.1 Crea dos listas:

- Una que se llame "valores" que tenga 7 números aleatorios.

- Otra que se llame "indices" que tenga 7 strings que se llamen como los días de la semana.

```python
valores = np.random.randint(0, 30, size=7).tolist()
indices = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

print(valores)
print(indices)
```

3.2 A partir de estas dos listas crea una Serie donde la variable "valores" sean los valores de la Serie y la variable indices sean los índices de la Serie.

```python
serie1 = pd.Series(data=valores, index=indices)
print(serie1)
```

3.3 A partir de las dos listas anteriores crea un diccionario usando un bucle for. Para hacer esto tenemos el método zip que nos permite iterar por dos listas a la vez. Más información sobre este método aquí.

```python
valores = [4, 22, 4, 9, 27, 0, 21]
indices = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

diccionario2 = {}

for dia, valor in zip(indices, valores):
    diccionario2[dia] = valor

print(diccionario2)
```

3.4 Crea una Serie con el diccionario creado en el punto anterior.

```python
serie2 = pd.Series(diccionario2)

print(serie2)
```
