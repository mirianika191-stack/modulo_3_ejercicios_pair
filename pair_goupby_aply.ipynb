1. Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
df = pd.read_csv("file/world_data_full.csv")
df.head()
```

```python
col_to_numeric = [
    "land_area", "armed_forces_size", "co2_emissions", "gdp", "population",
    "urban_population", "latitude", "longitude", "density", "cpi"
]

def convertir_a_numerico(df, columnas):
    for col in columnas:
        if col in df.columns:      
            df[col] = df[col].astype(str)
            df[col] = df[col].str.replace(",", "")
            df[col] = df[col].str.replace("%", "")
            df[col] = pd.to_numeric(df[col], errors="coerce")
        else:
            print("La columna '{}' no existe en el DataFrame.".format(col))
    return df

df = convertir_a_numerico(df, col_to_numeric)
```

5. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:

- Primero, necesitas calcular la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate".

- Luego, calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".

- Ahora, divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado.

- Encuentra el país que tiene el valor más alto en esta nueva columna calculada.

```python
tasa_impuestos_continente = df.groupby("Continent")["total_tax_rate"].mean()
print("Tasa de impuestos total promedio por continente:")
print(tasa_impuestos_continente)
```

```python
densidad_continente = df.groupby("Continent")["density"].mean()
print("\nDensidad poblacional promedio por continente:")
print(densidad_continente)
```

```python
df["tax_per_density"] = df["total_tax_rate"] / df["density"]
print("\nColumna 'tax_per_density' creada (primeras filas):")
print(df[["country", "Continent", "tax_per_density"]].head())
```

```python
pais_max_tax_density = df.loc[df["tax_per_density"].idxmax()]
print("\nPaís con la mayor tasa de impuestos por densidad poblacional:")
print(pais_max_tax_density[["country", "Continent", "tax_per_density"]])
```

8. ¿Cuál es la tasa de maternidad promedio por tasa de impuestos total? Para resolver el ejercicio deberás:

- Primero, calcula la tasa de maternidad promedio para cada continente. Puedes usar la columna "maternal_mortality_ratio" que representa la tasa de mortalidad materna.

- Calcula la tasa de impuestos total promedio para cada continente. Puedes usar la columna "total_tax_rate" que representa la tasa de impuestos total en cada continente.

- Luego, divide la tasa de maternidad promedio por la tasa de impuestos total promedio para obtener la relación entre ambas tasas.

- Encuentra el país con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total.

```python
maternal_promedio = df.groupby("Continent")["maternal_mortality_ratio"].mean()
print("Tasa de maternidad promedio por continente:")
print(maternal_promedio)
```

```python
tax_promedio = df.groupby("Continent")["total_tax_rate"].mean()
print("\nTasa de impuestos total promedio por continente:")
print(tax_promedio)
```

```python
relacion_mater_tax = maternal_promedio / tax_promedio
print("\nRelación entre tasa de maternidad y tasa de impuestos total por continente:")
print(relacion_mater_tax)
```

```python
df["maternal_per_tax"] = df["maternal_mortality_ratio"] / df["total_tax_rate"]


pais_max_mater_tax = df.loc[df["maternal_per_tax"].idxmax(), ["country", "Continent", "maternal_per_tax"]]
print("\nPaís con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total:")
print(pais_max_mater_tax)
```
